```python
client = StravaClient()

df_telemetria = client.get_activity_streams(activity_id=17691712197)
```

```python
df_telemetria.head()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Crear la figura principal habilitando un eje Y secundario
fig = make_subplots(specs=[[{"secondary_y": True}]])

# 2. Añadir la línea de Altitud al eje Y principal (izquierda)
# El modo 'lines' y la personalización del color ayudan a diferenciar claramente los datos
fig.add_trace(
    go.Scatter(
        x=df_telemetria["time"], 
        y=df_telemetria["altitude"], 
        name="Altitud (m)", 
        mode="lines", 
        line=dict(color="#1f77b4", width=2) # Color azul
    ),
    secondary_y=False,
)

# 3. Añadir la línea de Frecuencia Cardíaca al eje Y secundario (derecha)
fig.add_trace(
    go.Scatter(
        x=df_telemetria["time"], 
        y=df_telemetria["heartrate"], 
        name="Ritmo Cardíaco (bpm)", 
        mode="lines", 
        line=dict(color="#d62728", width=2) # Color rojo
    ),
    secondary_y=True,
)

# 4. Configurar nombres de la gráfica y comportamiento interactivo
fig.update_layout(
    title_text="Telemetría de Strava: Altitud vs Frecuencia Cardíaca",
    template="plotly_white", # Un tema limpio
    hovermode="x unified",    # Permite ver ambos valores exactos al pasar el cursor sobre cualquier punto del tiempo
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)

# 5. Nombrar el eje X
fig.update_xaxes(title_text="Tiempo Transcurrido (s)") 
# Nota: Si prefieres usar la hora real en lugar de los segundos transcurridos, cambia 'time' por 'timestamp' en el X de arriba.

# 6. Nombrar y colorear dinámicamente los ejes Y para que coincidan con las líneas
fig.update_yaxes(title_text="Altitud (metros)", secondary_y=False, title_font=dict(color="#1f77b4"), tickfont=dict(color="#1f77b4"))
fig.update_yaxes(title_text="Ritmo Cardíaco (bpm)", secondary_y=True, title_font=dict(color="#d62728"), tickfont=dict(color="#d62728"))

# ¡Renderizar la gráfica interactiva!
fig.show()
```
